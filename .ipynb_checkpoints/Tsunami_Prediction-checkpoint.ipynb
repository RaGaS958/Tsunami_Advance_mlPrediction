```python
df = df[df['dmin'] != 0]
Q1 = df['dmin'].quantile(0.25)
Q3 = df['dmin'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Compute mean of non-outlier values
mean_value = df.loc[(df['dmin'] >= lower_bound) & (df['dmin'] <= upper_bound), 'dmin'].mean()

# Replace outliers with mean
df['dmin'] = np.where((df['dmin'] < lower_bound) | (df['dmin'] > upper_bound),
                       mean_value, df['dmin'])
```

```python
plt.figure(figsize=(7,7))
sns.histplot(x=df['dmin'],kde=True)
```

```python
df_clean=df.copy()
```

```python
from sklearn.preprocessing import StandardScaler
cols=[  'sig', 'nst', 'gap', 'depth',
       'latitude', 'longitude']
scaler=StandardScaler()
df_clean[cols]=scaler.fit_transform(df_clean[cols])
df_clean
```

```python
df_clean.columns
```

```python
from scipy.stats import pearsonr

# ----------------------------------
# Pearson Correlation Calculation
# ----------------------------------

# List of features to check against target
selected_features = ['magnitude', 'cdi', 'mmi', 'sig', 'nst', 'dmin', 'gap', 'depth',
       'latitude', 'longitude', 'Year', 'Month']

correlations = {
    feature: pearsonr(df_clean[feature], df_clean['tsunami'])[0]
    for feature in selected_features
}
correlation_df = pd.DataFrame(list(correlations.items()), columns=['Feature', 'Pearson Correlation'])
correlation_df.sort_values(by='Pearson Correlation', ascending=False)
```

```python
from scipy.stats import chi2_contingency

alpha = 0.05

df_clean['Tsunami_bin'] = df_clean['tsunami']
chi2_results = {}

for col in cols:
    contingency = pd.crosstab(df_clean[col], df_clean['Tsunami_bin'])
    chi2_stat, p_val, _, _ = chi2_contingency(contingency)
    decision = 'Reject Null (Keep Feature)' if p_val < alpha else 'Accept Null (Drop Feature)'
    chi2_results[col] = {
        'chi2_statistic': chi2_stat,
        'p_value': p_val,
        'Decision': decision
    }

chi2_df = pd.DataFrame(chi2_results).T
chi2_df = chi2_df.sort_values(by='p_value')
chi2_df
```

```python
df_clean.drop(columns=['Tsunami_bin'],inplace=True)
```

# KNN

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=5)
```

```python
df_clean.columns
```

```python
from sklearn.model_selection import train_test_split

x=df_clean.drop(columns=['tsunami','Month','mmi','cdi','dmin','gap','depth','magnitude'],axis=1)
y=df_clean['tsunami']
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
x.columns
```

```python
knn.fit(X_train,y_train)
```

```python
y_knn_predict=knn.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, f1_score, classification_report

acc=accuracy_score(y_test,y_knn_predict)
acc*100
```

```python
from sklearn.metrics import confusion_matrix

confusion_matrix(y_test,y_knn_predict)
```

```python
print(classification_report(y_test,y_knn_predict))
```
